#1-Descripción del problema:

Entrada: un arreglo de tamaño n con datos desordenados

Salida: un arreglo de tamaño n ordenado

Captura.PNG

Existen muchos tipos de algoritmos para ordenar conjuntos de datos, cada cual con casos en los que el rendimiento puede verse severamente afectado, pero hay un método que mantiene la eficiencia de manera constante en cualquier caso posible, este es el Merge Sort


#2-Código:

```python
def merge(array, L, M, verbose = False, comps=[0]):
        comps[0] += 1

        i = j = k = 0
        while i < len(L) and j < len(M):
            comps[0] += 1
            if L[i] < M[j]:
                comps[0] += 1
                array[k] = L[i]
                i += 1
            else:
                array[k] = M[j]
                j += 1
            k += 1

        while i < len(L):
            comps[0] += 1
            array[k] = L[i]
            i += 1
            k += 1

        while j < len(M):
            comps[0] += 1
            array[k] = M[j]
            j += 1
            k += 1

        if verbose:
          print("Arreglo obtenido tras Merge:")
          print(array)

def mergeSort(array, verbose = False, calls = [0], comps = [0]):
    calls[0] += 1
    if len(array) > 1:

        r = len(array)//2
        L = array[:r]   
        M = array[r:]

        if(verbose):
          print("\nArreglo recibido por la iteración " + str(calls[0]))
          print(array)

          print("Sub arreglo obtenido 1")
          print(L)
          print("Sub arreglo obtenido 2")
          print(M)

        mergeSort(L, verbose, calls, comps)
        mergeSort(M, verbose, calls, comps)
        merge(array, L, M, verbose, comps)

  
def printList(array):
    for i in range(len(array)):
        print(array[i], end=" ")
    print()


if __name__ == '__main__':
    calls = [0]
    comps = [0]
    array = [6, 5, 12, 10, 9, 1]
    print("Arreglo de entrada: ")
    print(array)

    verbose = False
    mergeSort(array, verbose, calls, comps)

    print("\nArreglo de salida: ")
    print(array)

    if verbose:
      print("\nComparaciones realizadas por Merge")
      print(comps[0])

      print("\nLlamadas recursivas a MergeSort")
      print(calls[0]-2)
```

#2.1 Descripción del algoritmo

-El algoritmo recibe un arreglo de n posiciones.

-Este arreglo se divide en dos arreglos de tamaño n/2, teniendo cada uno una mitad opuesta de los datos del arreglo original.

-Estos arreglos y los subsecuentes pasan por el mismo proceso hasta quedar con n arreglos de una posición.

-Se utiliza la función merge, esta recibirá los arreglos de una posición resultantes y los combinará en el orden correcto.

#3-Tiempo de ejecución:

**Teorema**: El algoritmo merge sort tiene complejidad O(nlogn) para cualquier caso, debido a que este siempre divide el arreglo en dos mitades y siempre ocupa una cantidad de tiempo lineal para combinar dos mitades.

**Prueba del teorema**:

Cuando dividimos un número a la mitad de forma sucesiva, se representa con una función logarítmica, la cual es log n.

Luego, para combinar los sub arreglos resultantes de dividir el arreglo original de n elementos, se necesita un tiempo de ejecución de n.

Por lo tanto, el tiempo de ejecución del algoritmo en su conjunto nos da un tiempo de ejecución de O(nlogn)

#4.1-Corretitud de la función Merge

**Teorema**: La función merge recibe un arreglo de largo n [a1, a2, ... an] y otro de largo m [a1, a2, ... am], luego retorna un arreglo de largo n+m ordenado

**Prueba del teorema**: En cada iteración de la función merge se colocará cada elemento del par de arreglos que recibe de manera ordenada en un arreglo n+m, sí que para probar el teorema utilizaremos la propiedad de bucle invariante.

**Inicialización**

En la primera iteración el arreglo de salida contendrá solo un valor, el cual se puede considerar que está ordenado, de manera que la propiedad se cumple.

**Mantención**

El arreglo de salida contendrá i-1 (siendo i el número de iteraciones) datos ordenados, siendo todos los valores restantes a ordenar mayores que los elementos ya ordenados.

**Finalización**

Ya que se cumple la propiedad al inicio y durante la ejecución del algoritmo, se puede afirmar que se generará un arreglo [a1, a2, ... , a(n+m)] con los mismos elementos de los arreglos de entrada ordenados de menor a mayor.


#4.2-Correctitud de la función Merge Sort

**Teorema**:
El algoritmo merge sort recibe un arreglo de n elementos, y retorna un arreglo de n elementos ordenados.

**Prueba del teorema**: El algoritmo se llamará a si mismo de forma recursiva, dividiendo en cada iteración los arreglos de largo n que recibe en dos arreglos de largo n/2 con los mismos elementos, para ser combinados en el orden correcto por la función merge, a la cual ya probamos su correctitud.

**Inicialización**:

En la primera iteración se dividirá el arreglo original en dos arreglos [q...r]
y [k...p] siendo q y p los límites laterales del arreglo original, y r y k los elementos del medio.

**Mantención**:

Los arreglos subsiguientes se dividirán en arreglos cuya suma de elementos seguirá siendo la cantidad de elementos del arreglo dividido.

**Finalización**:

Ya que cada iteración del algoritmo mantiene la misma cantidad de datos, y que el funcionamiento de la función merge es correcta, se puede afirmar que de un arreglo de largo n, se obtendrá un arreglo de largo n ordenado.

```python
import matplotlib.pyplot as plt
import random

x=[n for n in range(5,20)] 

y=[] 
y1=[]
y2=[]

rec = [0]
for n in range(5,20):
  a = random.sample(range(1, 100), n)
  comps = [0]
  mergeSort(a, False, rec, comps)
  y.append(comps[0])

  c = []
  for i in range(0,n-1):
    c.append(i)

  comps3 = [0]
  mergeSort(c, False, rec, comps3)
  y2.append(comps3[0])

for n in range(5,20):
  b = random.sample(range(1, 100), n)
  b[0] = 99999
  b[n-1] = 88888 
  
  comps2 = [0]
  mergeSort(b, False, rec, comps2)
  y1.append(comps2[0])
  

plt.plot(x,y)
plt.plot(x,y1)
plt.plot(x,y2)
plt.legend(["MergeSort", "Peor caso teórico", "Mejor caso teórico"])

plt.xlabel('n')
plt.ylabel('Número de operaciones')
plt.show()
```

**Observaciones**:

En el gráfico se puede apreciar como el merge sort no varía su número de comparaciones en el peor ni en el mejor caso teórico, demostrando que su complejidad temporal es la misma para todas las situaciones.

#6.2 Comparación con insertionSort

```python
import matplotlib.pyplot as plt
import random

x=[n for n in range(5,20)] 
y=[] 
y2=[]

def insertionSort(numeros, verbose=False):
    comparaciones = 0
    for i in range(1, len(numeros)):
        if verbose == True: print("\nPasada",i)
        key = numeros[i] 
        j = i - 1
        while j >= 0 and key < numeros[j]:
            comparaciones+=1
            if verbose == True:print(numeros)
            numeros[j + 1] = numeros[j] 
            j = j - 1
        if verbose == True:print(numeros)
        numeros[j + 1] = key

        if verbose == True:print("\nAl finalizar la pasada: ") 
        if verbose == True:print(numeros)

    return comparaciones, verbose

for n in range(5,20):
  a = random.sample(range(1, 100), n)
  comps = [0]
  mergeSort(a, False, comps, comps)
  y.append(comps[0])

for n in range(5,20):
  a = random.sample(range(1, 100), n)
  comps2,e = insertionSort(a)
  y2.append(comps2)

plt.plot(x,y)
plt.plot(x,y2)
plt.legend(["MergeSort", "InsertionSort"])

plt.xlabel('n')
plt.ylabel('Número de operaciones')
plt.show()
```

**Observaciones**:

Se puede apreciar que el merge sort requiere una mayor cantidad de comparaciones a diferencia del insertion sort, pero también se puede apreciar como este último tiene mucha variación con respecto a las comparaciones, a diferencia del merge sort que mantiene de forma más o menos constante su gráfica.
