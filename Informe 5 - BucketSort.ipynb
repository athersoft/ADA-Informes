#1-Problema de ordenamiento 

Entrada: Secuencia de n números $[a1, a2, ... , an]$

Salida: Permutación ordenada de la secuencia de entrada de tal manera que $a1 < a2 < ... < an$

La necesidad de organizar elementos es un problema común que se presenta dentro de área informática, teniendo diversos métodos a elegir cada uno con sus propias ventajas y desventajas.

En esta situación, vamos a analizar el algoritmo bucket sort, este tiene la capacidad de mejorar el rendimiento de algoritmos de ordenamiento ya existentes.

#2-Código

```python
from termcolor import colored
import random

def insertionSort(b):
    for i in range(1, len(b)):
        up = b[i]
        j = i - 1
        while j >= 0 and b[j] > up: 
            b[j + 1] = b[j]
            j -= 1
        b[j + 1] = up     
    return b     
              
def bucketSort(x, verbose = False):
    arr = []
    slot_num = 10
    for i in range(slot_num):
        arr.append([])

    for j in x:
        index_b = int(slot_num * j) 
        arr[index_b].append(j)
      
    if verbose:
      print("\nBaldes desordenados")
      for i in range(0,10):
        if len(arr[i]) > 0:
          print("Balde", "{:.1f}".format(0.1*i), "-", "{:.1f}".format(0.1*(i+1)),":", arr[i])
        else:
          print("Balde", "{:.1f}".format(0.1*i), "-", "{:.1f}".format(0.1*(i+1)),": [Vacío]")

    for i in range(slot_num):
        arr[i] = insertionSort(arr[i])

    if verbose:
      print("\nBaldes ordenados")
      for i in range(0,10):
        if len(arr[i]) > 0:
          print("Balde", "{:.1f}".format(0.1*i), "-", "{:.1f}".format(0.1*(i+1)),":", arr[i])
        else:
          print("Balde", "{:.1f}".format(0.1*i), "-", "{:.1f}".format(0.1*(i+1)),": [Vacío]")

    k = 0
    for i in range(slot_num):
        for j in range(len(arr[i])):
            x[k] = arr[i][j]
            k += 1
    return x

arr = []

for i in range(0,15):
    a = round(random.uniform(0.001, 0.999),3)
    arr.append(a)

print("Arreglo recibido:", arr)


bucketSort(arr) 
print("\nArreglo resultante:", arr)
```

#2.1-Descripción del algoritmo

BucketSort.png

El bucketsort es un algoritmo muy útil cuando se tiene un conjunto de datos distribuido de manera uniforme, su funcionamiento se basa en dividir los elementos en arreglos más pequeños a los que llamaremos baldes, a los cuales se les aplicará un algoritmo de ordenamiento.

Su mayor ventaja es la de dividir los elementos de un arreglo en casos distintos, al aislar los casos, se evita que el algoritmo de ordenamiento principal deba trabajar con un arreglo desfavorable, en su lugar, debe trabajar con casos más pequeños potencialmente mejorando el tiempo de ejecución.

Para implementar el bucketsort se aplican los siguientes pasos:

1-Se divide los elementos del arreglo en intervalos, estos intervalos almacenarán datos en rangos de $0.1*i$ hasta $0.1*(i+1)$

2-Se crea un "balde" para cada intervalo, y se almacenan los elementos correspondientes a cada uno

3-A cada balde se le aplica un algoritmo de ordenamiento

4-Se juntan los elementos de los baldes ya ordenados

5-El arreglo ha sido ordenado

#2.2-Ejemplo de ejecución:

Para nuestro ejemplo utilizaremos el arreglo $A = [0.233, 0.111, 0.222]$

Se crearán 2 baldes con sus elementos en el orden original: $[0.111]$ y $[0.233, 0.222]$

Se aplicará un algoritmo de ordenamiento a ambos: $[0.111]$ y $[0.222, 0.233]$

Se juntan los baldes en un solo arreglo: $[0.111, 0.222, 0.233]$

El arreglo está ordenado!

#3-Correctitud

Teorema: 

El algoritmo bucketSort ordena correctamente un arreglo de n elementos

Prueba del teorema:

Para probar el teorema utilizaremos la inducción matemática.

Caso base:

Si el arreglo es de largo $1$ se retornará este mismo elemento.

Caso general:

Para una cantidad de datos $n$, se creará una cantidad acorde de baldes, luego, cuando estos se junten, se retornará un arreglo de $n$ datos ordenados, esto también se cumple para arreglos de largo n+1.


#4-Tiempo de ejecución:

Mejor caso:

Este se da cuando los datos ya están ordenados, de manera que en cada bucket creado se necesite un tiempo lineal, siendo n el tiempo requerido para crear los buckets y k para aplicar insertion sort a cada bucket, resultando en un tiempo de ejecución total de O(n+k)

Peor caso:

Si los elementos del arreglo están en un mismo rango, todos serán ubicados en un mismo bucket, y si además se encuentran en orden inverso, nos dará una complejidad temporal de O(n^2)

#5.1-Comparativa BucketSort vs RadixSort vs CountingSort

```python
import matplotlib.pyplot as plt
import datetime
from timeit import repeat
import random

x=[]; y=[]; x2=[]; y2=[]; x3=[]; y3=[]

def count_sort(arr):
    max_element = int(max(arr))
    min_element = int(min(arr))
    range_of_elements = max_element - min_element + 1
    count_arr = [0 for _ in range(range_of_elements)]
    output_arr = [0 for _ in range(len(arr))]
 
    for i in range(0, len(arr)):
        count_arr[arr[i]-min_element] += 1

    for i in range(1, len(count_arr)):
        count_arr[i] += count_arr[i-1]

    for i in range(len(arr)-1, -1, -1):
        output_arr[count_arr[arr[i] - min_element] - 1] = arr[i]
        count_arr[arr[i] - min_element] -= 1
 
    for i in range(0, len(arr)):
        arr[i] = output_arr[i]
 
    return arr

def countingSort(array, place):
    size = len(array)
    output = [0] * size
    count = [0] * 10

    for i in range(0, size):
        index = array[i] // place
        count[index % 10] += 1

    for i in range(1, 10):
        count[i] += count[i - 1]

    i = size - 1
    while i >= 0:
        index = array[i] // place
        output[count[index % 10] - 1] = array[i]
        count[index % 10] -= 1
        i -= 1

    for i in range(0, size):
        array[i] = output[i]

def radixSort(array):
    max_element = max(array)
    place = 1
    while max_element // place > 0:
        countingSort(array, place)
        place *= 10

      
for n in range(5,500):
  nums = []

  for i in range(0,n):
    a = round(random.uniform(0.001, 0.999),3)
    nums.append(a)

  t = repeat(setup="from __main__ import bucketSort", stmt=f"bucketSort({nums})", repeat=1, number=10)

  x.append(n)
  y.append(t)

for n in range(5,500):

  a = random.sample(range(1, 1000), n)

  t = repeat(setup="from __main__ import radixSort", stmt=f"radixSort({a})", repeat=1, number=10)

  x2.append(n)
  y2.append(t)

for n in range(5,500):

  b = random.sample(range(1, 1000), n)

  t = repeat(setup="from __main__ import count_sort", stmt=f"count_sort({b})", repeat=1, number=10)

  x3.append(n)
  y3.append(t)


plt.plot(x,y)
plt.plot(x2,y2)
plt.plot(x3,y3)
plt.legend(["Bucketsort", "RadixSort", "CountingSort"])

plt.xlabel('n')
plt.ylabel('time in ms')
plt.show()
```

Análisis de resultados:

En la gráfica se puede observar que si bien Bucketsort es más eficiente con arreglos más pequeños, este ve su rendimiento afectado negativamente conforme aumenta el tamaño del arreglo, esto se debe a que los baldes generados serán más largos, aumentando mucho el tiempo de ejecución del insertion sort.

Por otro lado, el counting sort resulta menos eficiente con arreglos de menor tamaño, pero su estabilidad provoca que conforme se alarga los arreglos de prueba, este resulte más eficiente que bucket sort y radix sort

#5.2-Tiempo de ejecución insertionSort vs bucketSort implementado con insertionSort

```python
import matplotlib.pyplot as plt
import datetime
from timeit import repeat

x=[]; y=[]; x2=[]; y2=[]

for n in range(5,500):
  nums = []
  nums2 = []

  for i in range(0,n):
    a = round(random.uniform(0.001, 0.999),3)
    nums.append(a)
    nums2.append(a)

  t = repeat(setup="from __main__ import bucketSort", stmt=f"bucketSort({nums})", repeat=1, number=10)
  t2 = repeat(setup="from __main__ import insertionSort", stmt=f"insertionSort({nums2})", repeat=1, number=10)

  x.append(n)
  y.append(t)

  x2.append(n)
  y2.append(t2)


plt.plot(x,y)
plt.plot(x2,y2)
plt.legend(["BucketSort", "InsertionSort"])

plt.xlabel('n')
plt.ylabel('time in ms')
plt.show()
```

Análisis de resultados:

En la gráfica se puede observar que a pesar de que el bucket sort está implementado utilizando insertion sort, el rendimiento es mucho mayor a si utilizamos el insertion sort por si solo.

Esto es debido a que al dividir los casos en rangos más pequeños, el insertion sort debe realizar menos comparaciones, además, se aíslan posibles casos desfavorables.
